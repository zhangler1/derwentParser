```python

sars=pd.read_csv("output\\"+filename,index_col=0)
```

```python
np.sum(sars.isnull() )
```

```python
sars=sars.dropna(how="all",axis=0,subset=["publishDate","language" ,"pages" ])#删除读取无用行
sars.head()
```

```python
sars.publishDate=sars.publishDate.fillna(sars.publishDate.mean())
sars.language=sars.language.fillna(sars.language.mode()[0])
sars.pages=sars.pages.fillna(method="bfill")
```
