```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
pa=pd.read_csv("output/machinepatents.csv",index_col=0)
pa.head(5)
```

```python
Aset=["JiangSu","World","China"] 
for region in Aset:
    data=pa[pa["region"]==region].groupby("publishDate").count()["PatentNumber"]
    data
    plt.figure(figsize=(6,6))
    plt.plot(data.index,data.values)
    for a,b in zip(data.index,data.values):
        plt.text(a,b*0.93 , '{} '.format(b), ha='center', va= 'bottom',fontsize=12)
    plt.savefig("专利随时间年限图"+region+".png")
```
